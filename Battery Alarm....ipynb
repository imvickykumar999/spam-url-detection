```python
# recently my lapi(in warrenty) ran into battery problem, so i had to go to service center...
# then i thought i should code an alarm which alert me when battery reach 97% then i could switch it off.
# while coding this programme i faced many errors and a time came when i thought i should leave this project,
# but i made hit and trials many experiments then finally i started working perfectly :)
    

import psutil, pyttsx3, time

battery = psutil.sensors_battery()
speaker = pyttsx3.init()

plugged = battery.power_plugged
percent = battery.percent

# maxbat = percent + 1
alarm = 60  # sec
maxbat = 97

if plugged == False:
    
    for i in range(5):
        speaker.say(str(percent) + '%' + '     Battery not Plugged in...')
        speaker.runAndWait()
        plugged = "Not Plugged In"
else:
    plugged = "Plugged In"

status = str(percent) + '% | ' + plugged
print(status)

speaker.say(status)
speaker.runAndWait()

count = 0
while True and plugged == "Plugged In":
    
    count = count + 1
    print(str(count), str(percent), str(maxbat), plugged, str(battery))
    time.sleep(10)

    battery = psutil.sensors_battery()
    pluggedbool = battery.power_plugged
    percent = battery.percent

    if pluggedbool == False:
        plugged = "Not Plugged In"
    else:
        plugged = "Plugged In"
    
    if percent >= maxbat:
        
        from pygame import mixer 
        mixer.init()
        
        mixer.music.load(r"C:\Users\Vicky Kumar\Music\Vicky_ My music\sun sonio - New Sad Song Best Heart Touching Video(MP3_128K).mp3") 
        mixer.music.set_volume(0.9) 
        mixer.music.play() 
        
        x = 'Battery percentage is GREATER than : {}'.format(maxbat)
        y = '    Do you wanna switch Off alarm ?'
        
        speaker.say(x + y)
        speaker.runAndWait()

        print(x)
        time.sleep(alarm)
        mixer.music.stop()
        break

#         yno = int(input(y))
#         if yno == 1:
#             mixer.music.stop()
#             break
    
    else:
        continue
    
```

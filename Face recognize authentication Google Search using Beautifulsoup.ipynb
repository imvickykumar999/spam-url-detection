```python
# one(1) time Execution

input('Write start : ')

# def initialise():

import cv2, os
import numpy as np
import webbrowser
import pyttsx3
import sys
from os import listdir
from os.path import isfile, join


speaker = pyttsx3.init()
speaker.say("Testing Speaker...")
speaker.runAndWait()

yourname = 'NULL'

# Load HAAR face classifier
face_classifier = cv2.CascadeClassifier('C:/Users/Vicky Kumar/Desktop/face detection attendence file/haarcascade_frontalface_default.xml')

# =========================================================================================================


# def takephoto():

speaker.say("Enter your name")
speaker.runAndWait()
yourname = input('Enter your name : ')

speaker.say("Hello, " + str(yourname))
speaker.runAndWait()

speaker.say('     Get ready and Please Smile, It\'s Selfie Time...')
speaker.runAndWait()

# import os mkdir str(yourname)
# os.mkdir(datasets + str(yourname) + '/')

datasets = 'C:/Users/Vicky Kumar/Desktop/face detection attendence file/model/'
# path = os.path.join(datasets, yourname) 
# if not os.path.isdir(path): 
#     os.mkdir(path) 


# Load functions
def face_extractor(img):
    # Function detects faces and returns the cropped face
    # If no face detected, it returns the input image

#   gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(img, 1.3, 5)

    if faces is ():
        return None

    # Crop all faces found
    for (x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face


def photo(ip, count, totalphoto):

    # Initialize Webcam of given ip
    cap = cv2.VideoCapture(ip)
        
    # Collect 100 samples of your face from webcam input
    while True:

        ret, frame = cap.read()
        if face_extractor(frame) is not None:
            count += 1
            face = cv2.resize(face_extractor(frame), (200, 200))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

            # Save file in specified directory with unique name
            file_name_path = 'C:/Users/Vicky Kumar/Desktop/face detection attendence file/model/'+ str(count) + '.jpg'
#              + str(yourname) + '/' 
            cv2.imwrite(file_name_path, face)

            # Put count on images and display live count
            cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
            cv2.imshow('Face Cropper', face)

        else:
            print("Face not found")
            pass

        if cv2.waitKey(1) == 13 or count == totalphoto: #13 is the Enter Key
            break 

            
cap = cv2.VideoCapture(0)

print('Selfie Time')
for _ in range(5):
    speaker.say("Selfie Time")
    speaker.runAndWait()
photo(0,0,100)

# print('ID Time')
# for _ in range(5):
#     speaker.say("ID Time")
#     speaker.runAndWait()
# # phone camera ip address = 0(for now) as arguments for id
# photo(0,100,200)


cap.release()
cv2.destroyAllWindows()
print("Collecting Samples Complete")


# =============================================================================================================


# print(cv2.__version__)

# os.chdir(r'C:\Users\Vicky Kumar\Desktop\face detection attendence file\model')

# for i in range(3):
#     print(os.listdir()[i])


# Get the training data we previously made
data_path = 'C:/Users/Vicky Kumar/Desktop/face detection attendence file/model/'
# + str(yourname) + '/'

# a=listdir('d:/faces')
# print(a)
# """
onlyfiles = [f for f in listdir(data_path) if isfile(join(data_path, f))]

# Create arrays for training data and labels
Training_Data, Labels = [], []

# Open training images in our datapath
# Create a numpy array for training data
for i, files in enumerate(onlyfiles):
    image_path = data_path + onlyfiles[i]
    images = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    Training_Data.append(np.asarray(images, dtype=np.uint8))
    Labels.append(i)

# Create a numpy array for both training data and labels
Labels = np.array(Labels, dtype=np.int32)
image=sys.argv[1]
img = cv2.imread(image)
face = face_classifier.detectMultiScale(img, 1.1, 4)
model = cv2.face.LBPHFaceRecognizer_create()

# Initialize facial recognizer
# model = cv2.face_LBPHFaceRecognizer.create()
# model=cv2.f
# NOTE: For OpenCV 3.0 use cv2.face.createLBPHFaceRecognizer()

# Let's train our model 
model.train(np.asarray(Training_Data), np.asarray(Labels))
print("Model trained sucessefully")


# ===========================================================================================

# execute this cell

from selenium import webdriver
import requests, time, pyttsx3, eyed3, cv2
import speech_recognition as sr
from pygame import mixer

face_classifier = cv2.CascadeClassifier('C:/Users/Vicky Kumar/Desktop/face detection attendence file/haarcascade_frontalface_default.xml')

yes = 1
while yes == 1:

    input('Are you READY ? (write yes) : ')
    
    # def recog():
    
    def face_detector(img, size=0.5):

        # Convert image to grayscale
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        if faces is ():
            return img, []

        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
            roi = img[y:y+h, x:x+w]
            roi = cv2.resize(roi, (200, 200))
        return img, roi


    # Open Webcam
    cap = cv2.VideoCapture(0)

    while(cv2.waitKey(30) != ord('q')):
    # while True:

        ret, frame = cap.read()
        image, face = face_detector(frame)

        try:
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

            # Pass face to prediction model
            # "results" comprises of a tuple containing the label and the confidence value
            results = model.predict(face)
            print(results)
            if results[1] < 500:
                confidence = int( 100 * (1 - (results[1])/400) )
                display_string = str(confidence) + '% Confident it is User'

            cv2.putText(image, display_string, (100, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 2)

            if confidence > 85 :
                cv2.putText(image, "Hey " + str(yourname), (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
                cv2.imshow('Face Recognition', image )
                speaker.say("         hii, i know you, you are " + str(yourname))
                speaker.runAndWait()
                cap.release()
                cv2.destroyAllWindows()

                # ==================================================================================

                # final project...

                r = sr.Recognizer()

                QUERY_FILE = (r"C:\Users\Vicky Kumar\Music\record\symptom of corona virus .wav ")
                HEY_GOOGLE_FILE = (r"C:\Users\Vicky Kumar\Music\record\hey google .wav") 

                with sr.AudioFile(QUERY_FILE) as qsource:
                    qaudio = r.record(qsource)

                with sr.AudioFile(HEY_GOOGLE_FILE) as hsource:
                    haudio = r.record(hsource)

                # Stop audio
                # mixer.music.stop()

                # with sr.Microphone() as qsource:
                #     print("Waiting to be called on as query...")
                #     haudio = r.listen(qsource)

                # with sr.Microphone() as hsource:
                #     print("Waiting to be called on as hey google...")
                #     qaudio = r.listen(hsource)

                try:
                    opt = int(input("Search Query as Audio(0) or User input(1) ? : "))
                    if opt == 0:

                        engine = pyttsx3.init() 
                        engine.say("You actually said : ") 
                        engine.runAndWait()

                        mixer.init()  
                        mixer.music.load(r"C:\Users\Vicky Kumar\Music\record\symptom of corona virus .mp3")  
                        mixer.music.set_volume(0.8)  
                        mixer.music.play()

                        duration = eyed3.load(r"C:\Users\Vicky Kumar\Music\record\symptom of corona virus .mp3").info.time_secs
                        time.sleep(duration)
    #                     time.sleep(3)

                        m = r.recognize_google(qaudio).lower()
                        print("Query Recognized as : " + m)

                        engine = pyttsx3.init() 
                        engine.say("Query Recognized as :    " + m) 
                        engine.runAndWait()

                    else:
                        m = input('Enter Query : ').lower()
                        print("Query Recognized as : " + m)

                        engine = pyttsx3.init() 
                        engine.say("Query Recognized as :    " + m) 
                        engine.runAndWait()

                    h = r.recognize_google(haudio).lower()
                    print("Audio file named, Hey Google Recognized as : " + h)
                #     text = input('Enter to check if exist : ')
                    text = "hey google"

                    if text in h:
                        engine = pyttsx3.init()
                        engine.say('yes, ' + text + ' exist in audio.wav as ' + h)
                        engine.runAndWait()
                        print('\nYes, ' + text + ' exist in audio.wav as ' + h)

                #         permission allowed to use google search engine

                        try: 
                            from googlesearch import search 
                        except ImportError as error: 
                            print("No module named 'google' found\n\n" + error) 

                        # string = input("Input the URL or string you want to search for : ")
                        string = m
                        query = string.replace(' ', '+')

                        browser = webdriver.Chrome(r'C:\Users\Vicky Kumar\Downloads\Compressed\chromedriver_win32\chromedriver.exe')

                        print("Click this link : ")
                        for j in search(query, tld="co.in", num=10, stop=1, pause=2): 
                            print(j)

                        engine = pyttsx3.init() 
                        engine.say("\nClick this link : " + j + " or      2 new chrome tab will open in new window...") 
                        engine.runAndWait()

                        r = requests.get(j)
                        x = r.content
                        print()

                        # print(x)

                        browser.get("https://www.google.com/search?q=" + query + "&start=0")

                        # Open a new window
                        browser.execute_script("window.open('');")

                        # Switch to the new window
                        browser.switch_to.window(browser.window_handles[1])
                        browser.get(j)

    #                     for i in range(2):
    #                         tab_index = int(input("Which Tab index you wanna close (0,1,2,...) ? : "))
    #                         browser.switch_to.window(browser.window_handles[tab_index])

    #                         seconds = int(input('Close tab after how much time (in seconds) : '))
    #                         time.sleep(seconds)
    #                         # close the active tab
    #                         browser.close()


                    else:
                        engine.say('no, ' + text + ' exist in audio.wav = ' + h)
                        print('\nNo, ' + text + ' exist in audio.wav = ' + h)

                except sr.UnknownValueError:
                    print("Could not understand audio")
                except sr.RequestError as e:
                    print("Could not request results; {0}".format(e))


                # ==================================================================================

                break

            else:
                cv2.putText(image, "Locked", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
                cv2.imshow('Face Recognition', image )
                speaker.say("Locked")
                speaker.runAndWait()
#                 cap.release()
#                 cv2.destroyAllWindows()
        except:
            cv2.putText(image, "No Face Found", (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.putText(image, "Locked", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Face Recognition', image )
    #         speaker.say("No face Found")
    #         speaker.runAndWait()
#             cap.release()
#             cv2.destroyAllWindows()
            pass
    yes = int(input('Wanna TRY Again ? (1/0) : '))
else:
    print("Thanks for USING this project...")
    speaker.say("Thanks for USING this project...")
    speaker.runAndWait()
```
